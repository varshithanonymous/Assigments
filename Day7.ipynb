```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import gensim
from gensim.utils import simple_preprocess

nltk.download('wordnet')

stop_words = set(stopwords.words('english'))

ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def process_text_with_gensim(text):
    """
    Process the input text using Gensim for basic preprocessing, including
    tokenization, stemming, and lemmatization.

    Args:
        text (str): The input text to process.

    Returns:
        dict: A dictionary with tokenized, stemmed, and lemmatized versions of the text.
    """
    # Convert text to lowercase and preprocess with Gensim's simple_preprocess
    tokens = simple_preprocess(text.lower())

    # Remove stopwords
    tokens_filtered = [word for word in tokens if word not in stop_words]

    # Apply stemming
    stemmed_words = [ps.stem(word) for word in tokens_filtered]

    # Apply lemmatization
    lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens_filtered]

    return {
        "original": tokens,
        "filtered": tokens_filtered,
        "stemmed": stemmed_words,
        "lemmatized": lemmatized_words
    }

if __name__ == "__main__":
    sample_text="Hi, im varshith Rallabandi m Roll No 429 of AIML-Epsilon, im doing this code as per the requirement of the day assignmetnS"

    processed_data = process_text_with_gensim(sample_text)

    print("Original Tokens:", processed_data["original"])
    print("Filtered Tokens:", processed_data["filtered"])
    print("Stemmed Tokens:", processed_data["stemmed"])
    print("Lemmatized Tokens:", processed_data["lemmatized"])
```
